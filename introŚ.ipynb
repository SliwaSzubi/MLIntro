```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}




d = pd.DataFrame(data=data)
d
```

```python
dCopy = d.copy()
dCopy.info()
```

```python
for col in ['size','color','gender','bought']:
    dCopy[col] = dCopy[col].astype('category')

dCopy['weight'] = dCopy['weight'].astype('float')

dCopy.info()
```

```python
dCopy.describe()
```

```python
dCopy.describe().T
```

```python
dCopy
```

```python
dCopy.describe(include=['category']).T
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(dCopy['bought'])
le.transform(dCopy['bought'])
```

```python
le.classes_
```

```python
dCopy['bought'] = le.fit_transform(dCopy['bought'])
dCopy
```

```python
dCopy['bought'] = le.inverse_transform(dCopy['bought'])
dCopy
```

```python
from sklearn.preprocessing import OneHotEncoder


encoder = OneHotEncoder(drop = 'first',sparse_output = False)
encoder.fit(dCopy[['size']])
```

```python
encoder.transform(dCopy[['size']])
```

```python
encoder.categories_
```
